```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import torch
import torch.nn as nn
import torch.nn.init as init
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from tqdm.auto import trange
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
if torch.cuda.is_available():
    device = torch.device("cuda:0")
else:
    device = torch.device("cpu")
print("Device:", device)
```

### Read Basket Option Data

```python
base_filename = 'basket_option100000x10'
```

```python
# Basket data stored in 10 files - load and concatenate
basket_data_list = list()
for i in range(0,10):
    new_basket_data = pd.read_csv(base_filename + str(i) + '.csv', index_col=[0])
    basket_data_list.append(new_basket_data) 
    
basket_data = pd.concat(basket_data_list)
basket_data.reset_index(inplace=True)
```

```python
drop_list = ['index','errorEstimate', 'samples', 'processTime']
basket_dataset = basket_data.drop(drop_list, axis=1)
basket_dataset.head()
```

### Split Data

```python
train_df = basket_dataset[:-40000]
test_df = basket_dataset[-40000:-20000]
validate_df = basket_dataset[-20000:]
```

```python
def prepareData(df):
    y = pd.DataFrame(df['optionValue'])
    X = df.drop(['optionValue'], axis=1)
    return X, y
```

```python
X_train, y_train = prepareData(train_df)
X_test, y_test = prepareData(test_df)
X_val, y_val = prepareData(validate_df)
```

### Standard Scaling

```python
standard_scalar = StandardScaler()
X_train = standard_scalar.fit_transform(X_train)
X_test = standard_scalar.transform(X_test)
X_val = standard_scalar.transform(X_val)
```

### Build Model

```python
def train_model(model, train_loader, test_loader, loss_fn, optimizer, epochs):
    train_errors, test_errors = [], []
    train_rmse_errors, test_rmse_errors = [], []

    tqdm_epoch = trange(epochs)
    for epoch in tqdm_epoch:
        model.train()
        train_loss, train_sq_sum = 0.0, 0.0

        # Training
        for batch_X, batch_y in train_loader:
            # Forward pass
            outputs = model(batch_X)
            loss = loss_fn(outputs, batch_y)
            train_sq_sum += torch.sum((outputs - batch_y) ** 2).item()
            
            # Backward pass and optimization
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            train_loss += loss.item() * batch_X.size(0)

        train_loss /= len(train_loader.dataset)
        train_rmse = np.sqrt(train_sq_sum / len(train_loader.dataset))
        train_errors.append(train_loss)
        train_rmse_errors.append(train_rmse)
        
        # Evaluation on test set
        model.eval()
        test_loss, test_sq_sum = 0.0, 0.0
        with torch.no_grad():
            for batch_X, batch_y in test_loader:
                outputs = model(batch_X)
                loss = loss_fn(outputs, batch_y)
                test_sq_sum += torch.sum((outputs - batch_y) ** 2).item()
                test_loss += loss.item() * batch_X.size(0)

        test_loss /= len(test_loader.dataset)
        test_rmse = np.sqrt(test_sq_sum / len(test_loader.dataset))
        test_errors.append(test_loss)
        test_rmse_errors.append(test_rmse)

        tqdm_epoch.set_description(f"Epoch {epoch+1}/{epochs} - Train loss: {train_loss:.4f}, \
            Test loss: {test_loss:.4f}, Train RMSE: {train_rmse:.4f}, Test RMSE: {test_rmse:.4f}")

    history = dict()
    history['train_loss'] = train_errors
    history['test_loss'] = test_errors
    history['train_rmse'] = train_rmse_errors
    history['test_rmse'] = test_rmse_errors

    return history
```

```python
no_epochs = 100
batch_size = 10000
```

```python
train_x = torch.Tensor(X_train).to(device)
train_y = torch.Tensor(y_train.to_numpy()).to(device)
train_dataset = TensorDataset(train_x, train_y)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=10000, 
                                           shuffle=True, drop_last=True)

val_x = torch.Tensor(X_val).to(device)
val_y = torch.Tensor(y_val.to_numpy()).to(device)
val_dataset = TensorDataset(val_x, val_y)
val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=1000, 
                                          shuffle=False, drop_last=True)
```

```python
class FeedForwardNN(nn.Module):
    def __init__(self):
        super(FeedForwardNN, self).__init__()
        self.hidden_layers = nn.ModuleList([nn.Linear(28, 200)] + [nn.Linear(200, 200) for _ in range(4)])
        self.output_layer = nn.Linear(200, 1)

    def forward(self, x):
        for layer in self.hidden_layers:
            x = torch.relu(layer(x))
        x = self.output_layer(x)
        return x
```

### MSE

```python
mse_model = FeedForwardNN().to(device)
```

```python
mse_loss_fn = nn.MSELoss()
mse_optimizer = optim.Adam(mse_model.parameters(), lr=0.001)
```

```python
mse_dict = train_model(mse_model, train_loader, val_loader,
                       mse_loss_fn, mse_optimizer, no_epochs)
```

### MAE

```python
mae_model = FeedForwardNN().to(device)
mae_loss_fn = nn.L1Loss()
mae_optimizer = optim.Adam(mae_model.parameters(), lr=0.001)

mae_dict = train_model(mae_model, train_loader, val_loader,
                       mae_loss_fn, mae_optimizer, no_epochs)
```

### Huber

```python
huber01_model = FeedForwardNN().to(device)
huber01_loss_fn = nn.HuberLoss(delta=0.1)
huber01_optimizer = optim.Adam(huber01_model.parameters(), lr=0.001)

huber01_dict = train_model(huber01_model, train_loader, val_loader,
                             huber01_loss_fn, huber01_optimizer, no_epochs)
```

```python
huber1_model = FeedForwardNN().to(device)
huber1_loss_fn = nn.HuberLoss(delta=1.0)
huber1_optimizer = optim.Adam(huber1_model.parameters(), lr=0.001)

huber1_dict = train_model(huber1_model, train_loader, val_loader,
                          huber1_loss_fn, huber1_optimizer, no_epochs)
```

```python
huber10_model = FeedForwardNN().to(device)
huber10_loss_fn = nn.HuberLoss(delta=10.0)
huber10_optimizer = optim.Adam(huber10_model.parameters(), lr=0.001)

huber10_dict = train_model(huber10_model, train_loader, val_loader,
                          huber10_loss_fn, huber10_optimizer, no_epochs)
```

### Log-Cosh

```python
class LogCoshLoss(nn.Module):
    def __init__(self):
        super(LogCoshLoss, self).__init__()

    def forward(self, y_pred, y_true):
        loss = torch.log(torch.cosh(y_pred - y_true))
        return loss.mean()
```

```python
logcosh_model = FeedForwardNN().to(device)
logcosh_loss_fn = LogCoshLoss()
logcosh_optimizer = optim.Adam(logcosh_model.parameters(), lr=0.001)

logcosh_dict = train_model(logcosh_model, train_loader, val_loader,
                          logcosh_loss_fn, logcosh_optimizer, no_epochs)
```

### Plot Training Loss

```python
epochs = range(1, no_epochs + 1)
```

```python
plt.figure(figsize=(7,5)) 
    
loss_values_mse = mse_dict['train_rmse']
loss_values_mae = mae_dict['train_rmse']
loss_values_huber01 = huber01_dict['train_rmse']
loss_values_huber1 = huber1_dict['train_rmse']
loss_values_huber10 = huber10_dict['train_rmse']
loss_values_logcosh = logcosh_dict['train_rmse']
    
plt.plot(epochs, loss_values_mse, color='k', linestyle='-', label = 'MSE')
plt.plot(epochs, loss_values_mae, color='k', linestyle='-', label = 'MAE')
plt.plot(epochs, loss_values_huber01, color='k', linestyle='--', label = 'Huber 01')
plt.plot(epochs, loss_values_huber1, color='k', linestyle='-.', label = 'Huber 1')
plt.plot(epochs, loss_values_huber10, color='k', linestyle=':', label = 'Huber 10')
plt.plot(epochs, loss_values_logcosh, color='k', linestyle=(0, (1, 5)), label = 'Log-Cosh')

#plt.title('Training Loss')
plt.xlabel('Epochs')
plt.ylabel('RSME')
plt.ylim(0.5,1.5)
plt.legend()
plt.savefig('TestBOLossFunction.png', dpi=300, bbox_inches='tight')
```

### Plot Validation Loss

```python
def smooth_curve(points, factor=0.9):
    smoothed_points = []
    for point in points:
        if smoothed_points:
            previous = smoothed_points[-1]
            smoothed_points.append(previous * factor + point * (1 - factor))
        else:
            smoothed_points.append(point)
    return smoothed_points
```

```python
loss_values_mse = smooth_curve(mse_dict['train_rmse'])
loss_values_mae = smooth_curve(mae_dict['train_rmse'])
loss_values_huber01 = smooth_curve(huber01_dict['train_rmse'])
loss_values_huber1 = smooth_curve(huber1_dict['train_rmse'])
loss_values_huber10 = smooth_curve(huber10_dict['train_rmse'])
loss_values_logcosh = smooth_curve(logcosh_dict['train_rmse'])
    
plt.plot(epochs, loss_values_mse, color='k', linestyle='-', label = 'MSE')
plt.plot(epochs, loss_values_mae, color='k', linestyle='-', label = 'MAE')
plt.plot(epochs, loss_values_huber01, color='k', linestyle='--', label = 'Huber 01')
plt.plot(epochs, loss_values_huber1, color='k', linestyle='-.', label = 'Huber 1')
plt.plot(epochs, loss_values_huber10, color='k', linestyle=':', label = 'Huber 10')
plt.plot(epochs, loss_values_logcosh, color='k', linestyle=(0, (1, 5)), label = 'Log-Cosh')
#plt.title('Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('RMSE')
plt.ylim(0.5,1.5)
plt.legend()
plt.savefig('TestBOLossFunctionValidation.png', dpi=300, bbox_inches='tight')
```
