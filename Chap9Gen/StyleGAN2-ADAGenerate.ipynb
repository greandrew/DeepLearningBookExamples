```python
def show_tensor_images(image_tensor, 
                       num_images=64, 
                       nrow=8,
                       filename=""):
    image_grid = make_grid(image_tensor[:num_images].detach().cpu(), nrow=nrow)
    img = image_grid.permute(1, 2, 0).squeeze()
    plt.axis('off')
    plt.imshow(img)
    if len(filename) > 0:
        plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()
```
