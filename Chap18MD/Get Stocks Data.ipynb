```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
```

### Download and preprocess the time series data

```python
# Define the list of stocks
stocks = ['AAPL', 'MSFT', 'IBM', 'WMT', 'JNJ', 'PG', 'XOM', 'BA',
          'KO', 'MCD', 'MMM', 'CAT', 'DIS', 'HPQ', 'MRK', 'PFE', 
          'NVDA', 'HD', 'INTC', 'AA', 'MO', 'T', 'GE', 'DD', 'CVX', 
          'JPM', 'AXP', 'C', 'BAC', 'GS', 'WFC', 'USB']

# Download the stock prices
data = yf.download(stocks, start='2001-01-02', end='2021-01-02')

# Select only the "Close" prices
close_prices = data['Close']

# Drop any rows with missing values
close_prices.dropna(inplace=True)

# Calculate the log returns
log_returns = np.log(close_prices / close_prices.shift(1))

# Drop any rows with missing values
log_returns.dropna(inplace=True)

# Define the window size
window_size = 60  # Change this to your desired window size

# Calculate the rolling correlation matrix
rolling_corr = log_returns.rolling(window_size).corr()
rolling_corr.dropna(inplace=True)

# Save the rolling correlation matrix to a CSV file
rolling_corr.to_csv('rolling_corr.csv')
```

```python
rolling_corr
```

```python
# Define the date
date = '2020-01-02'  # Change this to your desired date

# Extract the correlation matrix for the given date
corr_matrix = rolling_corr.loc[date]

# Convert the correlation matrix to a numpy array
corr_matrix_np = corr_matrix.values

print(corr_matrix_np)
```

```python
def plot_correlation_heatmap(correlation_matrix):
    fig, ax = plt.subplots()
    heatmap = ax.imshow(correlation_matrix, cmap='RdYlBu_r')
    cbar = plt.colorbar(heatmap)
    cbar.set_label('Correlation')

    # Set title and axis labels
    ax.set_title("Correlation Matrix Heatmap")

    # Show the plot
    plt.tight_layout()
    plt.show()
```

```python
plot_correlation_heatmap(corr_matrix_np)
```
