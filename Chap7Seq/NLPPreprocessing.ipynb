## NLP Preprocessing
This notebook provides a series of examples of common preprocessing steps for NLP. The examples use the Natural Language Tooklit `nltk` Python library.

```python
import nltk
```

### Tokenization
Tokenization is a fundamental step in Natural Language Processing (NLP) where text is broken down into smaller units called tokens.

```python
from nltk.tokenize import word_tokenize

# Download the necessary NLTK models (if not already downloaded)
nltk.download('punkt_tab')
```

```python
# Example sentence
sentence = "Natural Language Processing with Python is fun!"

# Tokenizing the sentence into words
tokens = word_tokenize(sentence)

print(tokens)
```

### Lowercasing
Lowercasing in NLP is a preprocessing step where all the text in a given dataset is converted to lowercase. 

```python
# Example sentence
sentence = "Natural Language Processing with Python is Exciting!"

# Converting the sentence to lowercase
lowercased_sentence = sentence.lower()

print(lowercased_sentence)
```

### Removing Punctuation, Numbers and Special Characters
Removing punctuation, numbers, special characters, and emojis is a common preprocessing step in NLP to simplify and standardize text data

```python
import string
import re

# Example text with punctuation, numbers, special characters, and emoji
text = "Hello, World! This is a test text with numbers 123, special characters @#$ and emoji 😀👍."

# Define a regex pattern for matching emojis
emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
```

```python
# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))
print(text)
```

```python
# Remove numbers
text = re.sub(r'\d+', '', text)
print(text)
```

```python
# Remove special characters and emojis
text = re.sub(r'[^A-Za-z0-9 ]+', '', text)
print(text)
```

### Stop Word Removal
Stop words removal is an essential preprocessing step in many NLP tasks, where frequently occurring words that carry minimal semantic importance are eliminated from the text. 

```python
from nltk.corpus import stopwords

# Download stop words set
nltk.download('stopwords')

# Example text
text = "This is a sample sentence, demonstrating the removal of stop words."
```

```python
# Tokenize the text
tokens = word_tokenize(text)

# Get English stop words
stop_words = set(stopwords.words('english'))

# Remove stop words
filtered_sentence = [word for word in tokens if not word in stop_words]

print(filtered_sentence)
```

### Removing Handles and URLs
Removing handles (commonly used in social media usernames) and URLs is a significant preprocessing step in text analysis, especially when dealing with data from social media platforms like X(Twitter) or Facebook.

```python
# Example text with a handle and a URL
text = "This is a tweet by @user, and it contains a URL https://example.com"

# Remove handles - anything that starts with @ followed by any number of characters
text_without_handles = re.sub(r'@\S+', '', text)
print(text_without_handles)
```

```python
# Remove URLs - a simple regex for common URL patterns
text_without_handles_urls = re.sub(r'http\S+|www.\S+', '', text_without_handles)
print(text_without_handles_urls)
```

### Stemming
Stemming in NLP refers to the process of reducing words to their base or root form, a technique that aims to consolidate different forms of a word into a common representation. 

```python
from nltk.stem import PorterStemmer

# Initialize the Porter Stemmer
porter = PorterStemmer()
```

```python
# Example words
words = ["running", "runs", "run", "runner", "easily", "fairly"]

# Stemming the words
stemmed_words = [porter.stem(word) for word in words]

print(stemmed_words)
```

### Lemmatization
Lemmatization in NLP is a sophisticated technique that involves reducing words to their base or dictionary form, known as the lemma.

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer

# Initialize the WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

# Example sentence
sentence = "The leaves are falling and the cats are running around."

# Tokenize the sentence
tokens = word_tokenize(sentence)

# Lemmatize each word
lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens]

print(lemmatized_words)
```
