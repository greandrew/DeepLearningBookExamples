```python
import pandas as pd
from transformers import pipeline
from datasets import load_dataset
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sentiment_analysis = pipeline("sentiment-analysis", model="Sigma/financial-sentiment-analysis")
```

```python
dataset = load_dataset("financial_phrasebank", "sentences_allagree")
```

```python
def pipeline_label_to_number(label):
    if label == "LABEL_0":
        return 0
    elif label == "LABEL_1":
        return 1
    elif label == "LABEL_2":
        return 2
```

```python
# Run the pipeline and collect predictions
predictions = []
ground_truths = []
for example in dataset["train"]:
    # Run sentiment analysis
    result = sentiment_analysis(example["sentence"])[0]
    predicted_label = pipeline_label_to_number(result["label"])
    predictions.append(predicted_label)

    # Ground truth label
    ground_truth_label = example["label"]
    ground_truths.append(ground_truth_label)
```

```python
# Generate confusion matrix
conf_matrix = confusion_matrix(ground_truths, predictions)
```

```python
filename = 'financial_phrasebank_sent.png'
```

```python
def plot_confusion(conf_matrix, filename):
    labels = ["Negative", "Neutral", "Positive"]
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels)
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.savefig(filename, dpi=300, format='png')
    plt.show()
```

```python
plot_confusion(conf_matrix, filename)
```
