# Load the dataset and plot the empirical distributions

```python
import pandas as pd
```

```python
filename_root = 'test_basket_mt'
```

```python
df1 = pd.concat(
    [pd.read_csv(f"{filename_root}{i}.csv") for i in range(20)],
    ignore_index=False
)

# Columns that need cleaning
cols_to_fix = ["maturity", "option_value", "error_estimate", "process_time", "samples"]

for col in cols_to_fix:
    if col in df1.columns:
        df1[col] = (
            df1[col]
            .astype(str)              # ensure string
            .str.strip("[]")         # remove brackets
            .astype(float)           # convert to numeric
        )
```

```python
import matplotlib.pyplot as plt

def plot_empirical_distribution(df, column_name, filename, dpi=300, color='blue'):
    """
    Plots the empirical distribution of a specified column in a pandas DataFrame.

    Parameters:
    df (DataFrame): The pandas DataFrame containing the data.
    column_name (str): The name of the column to plot.
    filename (str): The name of the file to save the plot to.
    dpi (int, optional): The resolution in dots per inch of the saved plot. Default is 300.
    """
    # Check if the column exists in the DataFrame
    if column_name not in df.columns:
        raise ValueError(f"The column '{column_name}' does not exist in the DataFrame.")
        
    # Plotting the empirical distribution
    plt.figure(figsize=(10, 6))
    df[column_name].plot(kind='hist', density=True, bins=100, alpha=0.5, color=color)
    
    # Plot formatting
    plt.xlabel(column_name)
    plt.ylabel('Density')
    
    # Save the plot to the specified file
    plt.savefig(filename, dpi=dpi)
    plt.show() 

    return f"Plot saved to {filename} at {dpi} dpi."
```

```python
plot_empirical_distribution(df1, 'maturity', 'maturity.png')
```
