### MNIST Edge Detection

In this notebook we will use convolution with the Scharr filter to demonstrate edge detection using images from the MNIST dataset.

```python
import torch
from torchvision import datasets, transforms
from scipy import signal
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

#### Load MNIST

```python
# Define a transform to normalize the data
transform = transforms.Compose([transforms.ToTensor()])

# Download and load the training data
train_set = datasets.MNIST('~/.pytorch/MNIST_data/', download=True, train=True, transform=transform)

# Download and load the test data
test_set = datasets.MNIST('~/.pytorch/MNIST_data/', download=True, train=False, transform=transform)
```

#### Plot MNIST examples

We first plot 8 examples from the MNIST training set.

```python
f, ax = plt.subplots(2, 4, figsize=(12,5), squeeze=True)

for r in range(2):
    for c in range(4):
        image_index = r * 4 + c
        ax[r,c].axis("off")
        ax[r,c].imshow(train_set.data[image_index], cmap='gray')
        
plt.savefig('MNISTNumbers.png', dpi=300)
```

#### Convolve with the Scharr Filter and Plot the Edges

The Scharr filter is a $3 \times 3$ tensor which can be used as an edge detector,
\begin{equation}
G = \begin{bmatrix}
3 & 0 & -3\\
10 & 0 & -10\\
3 & 0 & -3\\
\end{bmatrix}.
\end{equation}
To use it we perform a 2D convolution with the MNIST image using the <code>convolve2d</code> function from the signal processing package in <code>scipy</code>. Once convolved the same set of eight images are reprinted. The edges of the numbers are clearly visible.

```python
f, ax = plt.subplots(2, 4, figsize=(12,5), squeeze=True)

scharr = np.array([[ -3-3j, 0-10j,  +3 -3j],
                   [-10+0j, 0+ 0j, +10 +0j],
                   [ -3+3j, 0+10j,  +3 +3j]])

for r in range(2):
    for c in range(4):
        image_index = r * 4 + c
        ax[r,c].axis("off")
        grad = signal.convolve2d(train_set.data[image_index], scharr, boundary='symm', mode='same')
        ax[r,c].imshow(np.absolute(grad), cmap='gray')
        
plt.savefig('MNISTNumbersEdges.png', dpi=300)
```
