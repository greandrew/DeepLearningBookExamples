## TD Prediction in Taxi Environment

```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
import image
from tqdm.notebook import trange
import seaborn as sns
```

```python
env = gym.make('Taxi-v3', render_mode="rgb_array")
gamma = 0.99
alpha = 0.1
num_episodes = 1000
```

```python
env.reset()
frame = env.render()
```

```python
plt.imshow(frame)
plt.savefig('taxi_env.png', dpi=300)
```

```python
def td0_prediction(env, num_episodes):
    V = np.zeros(env.observation_space.n)

    for episode in trange(num_episodes):
        state = env.reset()[0]
        done = False
    
        while not done:
            action = env.action_space.sample()
            next_state, reward, done, _, _ = env.step(action)
            V[state] += alpha * (reward + gamma * V[next_state] - V[state])
            state = next_state
    return V
```

```python
V = td0_prediction(env, num_episodes)
```

```python
V = V.reshape((5,5,5,4))
```

```python
def plot_value_function(V, pass_loc, dest_loc, filename, cmap='coolwarm'):
    value_grid = V[:,:,pass_loc, dest_loc]
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(value_grid, annot=True, cmap=cmap, cbar=True, linewidths=.5)
    plt.xlabel('Taxi Column')
    plt.ylabel('Taxi Row')
    plt.savefig(filename, dpi=300)
    plt.show()
```

```python
plot_value_function(V, 4, 0, 'tdo_predvf.png')
```
