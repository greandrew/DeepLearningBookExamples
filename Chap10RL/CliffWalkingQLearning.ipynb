## Cliff Walking with Q-Learning

```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
import image
from tqdm.notebook import trange
import seaborn as sns
```

```python
# Initialize the environment
env = gym.make('CliffWalking-v1', render_mode='rgb_array')
```

```python
env.reset()
frame = env.render()
```

```python
plt.imshow(frame)
plt.tight_layout()
plt.savefig('cliff-walking.png', dpi=300)
```

```python
# Q-learning parameters
alpha = 0.1  # learning rate
gamma = 0.99  # discount factor
epsilon = 0.1  # exploration rate
num_episodes = 1000000  # number of episodes
```

```python
def epsilon_greedy(state, Q, epsilon):
    if np.random.rand() < epsilon:
        return env.action_space.sample()
    else:
        return np.argmax(Q[state])
```

```python
def q_learning(env, num_episodes, epsilon=0.1):
    # Initialize the Q-table
    Q = np.zeros((env.observation_space.n, env.action_space.n))
    # Training the agent using Q-learning
    for episode in trange(num_episodes):
        state = env.reset()[0]
        done = False
    
        while not done:
            action = epsilon_greedy(state, Q, epsilon)
            next_state, reward, done, _, _ = env.step(action)
            next_action = np.argmax(Q[next_state])
            Q[state, action] += alpha * (reward + gamma * Q[next_state, next_action] - Q[state, action])
            state = next_state
    return Q
```

```python
Q = q_learning(env, num_episodes)
```

```python
# Plotting the optimal policy
def plot_optimal_policy(Q, filename, cmap='coolwarm'):
    optimal_actions = np.argmax(Q, axis=1).reshape((env.unwrapped.shape[0], env.unwrapped.shape[1]))
    action_symbols = np.array(['↑','→', '↓', '←' ])  # Mapping actions to symbols
    policy_grid = action_symbols[optimal_actions]
    state_values = np.max(Q, axis=1).reshape((env.unwrapped.shape[0], env.unwrapped.shape[1]))
    
    plt.figure(figsize=(10, 7))
    for i in range(env.unwrapped.shape[0]):
        for j in range(env.unwrapped.shape[1]):
            plt.text(j, i, policy_grid[i, j], ha='center', va='center', fontsize=12)
    
    im = plt.imshow(state_values, cmap=cmap, interpolation='none')
    cbar = plt.colorbar(im, orientation='horizontal', pad=0.1)
    cbar.set_label('State Value')
    plt.title('Optimal Policy for CliffWalking-v0')
    plt.xlabel('Column')
    plt.ylabel('Row')
    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    plt.show()
```

```python
plot_optimal_policy(Q, 'cliffwalk_op.png')
```
